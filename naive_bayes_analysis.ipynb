```python
import pandas as pd
import numpy as np
import os
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import GridSearchCV
from sklearn.decomposition import PCA
import umap.umap_ as umap
import random
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.pipeline import Pipeline
from scipy.stats import randint, uniform
from sklearn.naive_bayes import GaussianNB

```

```python
data_dir = './GSE61260/'
```

```python
df = pd.read_csv(os.path.join(data_dir, 'combined.csv'))

df.head()
```

```python
le = LabelEncoder()

y = le.fit_transform(df['Disease'])

print(le.classes_)

X = df.drop(columns = ['Sample_ID', 'Disease'])
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)

print(X_train.shape)
print(X_test.shape)
```

```python
pipeline_steps = [
    ('scaler', StandardScaler()),              
    ('pca', PCA()), 
    ('classifier', GaussianNB()) 
]

pipeline = Pipeline(pipeline_steps)
print("\nDefined Pipeline:")
print(pipeline)
```

```python
parameters = {
    # Hyperparameters for PCA
    'pca__n_components': randint(2, min(X_train.shape[0] - 1, X_train.shape[1]) + 1),

    # Hyperparameters for Gaussian Naive Bayes
    'classifier__var_smoothing': np.logspace(-9, -2, 100) # e.g., from 1e-9 to 1e-2
}


random_search_pipeline = RandomizedSearchCV(
    estimator=pipeline,                
    param_distributions=parameters,  
    n_iter=50,                        # Increased iterations for better search
    cv=5,                             # 5-fold cross-validation
    scoring='f1_weighted',            # Recommended for potential class imbalance
    verbose=1,                         
    random_state=42,                  # Changed random_state for consistency
    n_jobs=-1                          
)

print("\nStarting RandomizedSearchCV on Pipeline (GaussianNB)...")
random_search_pipeline.fit(X_train, y_train)
print("RandomizedSearchCV on Pipeline complete.")

# --- 5. Retrieve the Best Pipeline and Evaluate ---
print("\nBest hyperparameters found for the pipeline:")
print(random_search_pipeline.best_params_)

print("\nBest cross-validation score:")
print(random_search_pipeline.best_score_)

best_pipeline_model = random_search_pipeline.best_estimator_
print("\nBest Pipeline Model:")
print(best_pipeline_model)

# Evaluate the best pipeline on the unseen test set
y_pred_tuned = best_pipeline_model.predict(X_test)

print("\n--- Evaluation of Best Tuned Pipeline on Test Set (GaussianNB) ---")
print(f"Accuracy: {accuracy_score(y_test, y_pred_tuned):.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_tuned, target_names=le.classes_))

conf_matrix = confusion_matrix(y_test, y_pred_tuned)
print("\nConfusion Matrix:")
sns.heatmap(conf_matrix)
```
