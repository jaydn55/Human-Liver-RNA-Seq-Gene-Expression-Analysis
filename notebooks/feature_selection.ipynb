```python
import pandas as pd
import os
```

```python
data_dir = '../data/'
```

```python
df_bio = pd.read_csv(os.path.join(data_dir, 'GSE61260.tsv'), sep='\t')

df_bio_t = df_bio.set_index('Gene').T
df_bio_t.index.name = 'Sample_ID'
df_bio_t = df_bio_t.reset_index()
df_bio_t.head()

test = df_bio_t.drop(columns = ['Sample_ID'])
test.describe()
```

```python
df_meta = pd.read_csv(os.path.join(data_dir, 'metadata_GSE61260.tsv'), sep='\t')
df_meta.head()
```

```python
df_combined = pd.merge(df_bio_t, df_meta[['refinebio_accession_code', 'refinebio_disease', 'characteristics_ch1_Sex', 'characteristics_ch1_age', 'characteristics_ch1_bmi']],
                           left_on='Sample_ID', right_on='refinebio_accession_code', how='inner').drop(columns=['refinebio_accession_code'])
df_combined.rename(columns = {"refinebio_disease" : "Disease"}, inplace=True)
df_combined = pd.get_dummies(df_combined, columns=['characteristics_ch1_Sex'], drop_first=True, dtype='int')
df_combined.head()

df_combined['Disease'].value_counts()
```

```python

df_combined = df_combined.drop(columns = ['characteristics_ch1_age', 'characteristics_ch1_Sex_male'])

path = os.path.join(data_dir, 'combined.csv')

df_combined.to_csv(path, index=False)

df_combined.head()
```
